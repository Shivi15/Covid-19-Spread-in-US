```python
import numpy as np
import pandas as pd

covid = pd.read_csv('us-counties.csv') # CSV data contains Covid 19 US counties
covid.head()
```

```python
covid.info()
```

```python
cpop = pd.read_excel('co-est2019-annres.xlsx',skiprows=2)
```

```python
cpop.head()
```

```python

# ignore all the explanations of the dataset at the very bottom
cpop = cpop[['Geographic Area','Population Estimate (as of July 1)']][0:3142]
cpop.tail(10)
```

```python
cpop_new = cpop[cpop['Geographic Area'].notnull()]
cpop_new['County'] = cpop_new['Geographic Area'].apply(lambda x: x.split(', ')[0][1:])
cpop_new['County'] = cpop_new['County'].apply(lambda x: x.split(' County')[0]) # for everywhere except AK + LA
cpop_new['County'] = cpop_new['County'].apply(lambda x: x.split(' Parish')[0]) # for LA
cpop_new['County'] = cpop_new['County'].apply(lambda x: x.split(' City')[0]) # for VA
cpop_new['State'] = cpop_new['Geographic Area'].apply(lambda x: x.split(', ')[-1])
cpop_new = cpop_new.iloc[1:]
cpop_new.head(70)
```

```python

# Test to see if Alaska works now
cpop_new[cpop_new['State']=='Alaska']
```

```python

# Test to see if Louisiana works now
cpop_new[cpop_new['State']=='Louisiana']
# 3 counties that all start with "West" have been combined, don't want to bother fixing that.
```

```python

#covid[covid['county']=='New York City']['county'] = 'New York'
#covid[covid['county']=='New York City']['county'].apply(lambda x: x.split(' '))
covid.loc[covid['county'] == 'New York City','county'] = 'New York'
covid[covid['state']=='New York']
```

```python
df = pd.merge(covid,cpop_new,left_on=['county','state'],right_on=['County','State'],how='inner')
df
```

```python

df.columns
```

```python
dfm = df[['date','County','State','cases','deaths','Population Estimate (as of July 1)']]
dfm.tail(10)
```

```python
dfm.to_excel('covid-us-county-population.xlsx')
```
